**SIMULATION EXERCISES**

13) **(1 PT)** Draw 100 random values from a standard normal distribution, using SEED = 314.

```python
from numpy.random import seed
from numpy.random import normal

seed(314)

dev = normal(loc=0, scale=1, size=100)

print(dev)

# compute cumulative sum
# https://mathworld.wolfram.com/CumulativeSum.html
# [INSERT CODE]
ccs = np.cumsum(dev)
print (ccs)

# plot the cumulative sum (called a random walk)
# [INSERT CODE]

plt.plot(ccs)
plt.title("a radom walk")
plt.show()
```

14) **(1 PT)** Write code to generate and plot 10 random walks. For reproducibility, explicitly use seed=314.  
Hint: Calling *plt.plot()* in a *for loop* will plot multiple lines on the same graph. For guidance on using a for loop, refer to the [lecture note on data types](https://github.com/youmilab/hudm5001/blob/main/lecture_notes/python/data_types.ipynb). 

```python
from numpy.random import seed
from numpy.random import normal

seed(314)

for _ in range(10): 
    dev = normal(loc=0, scale=1, size=100)
    ten_random_walk = np.cumsum(dev)
    plt.plot(ten_random_walk)
    plt.title("Ten Radom Walks")
plt.show()
```
